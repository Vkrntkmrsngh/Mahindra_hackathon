```python
import numpy as np
import pandas as pd
```

```python
df_train=pd.read_csv('train.csv',index_col='reservation_id')
df_test=pd.read_csv('test.csv',index_col='reservation_id')
```

```python
df_test['amount_spent_per_room_night_scaled']=-999
```

```python
#append train and test data
df=pd.concat([df_train,df_test],axis=0,sort=False)
```

```python
df['booking_date']=pd.to_datetime(df['booking_date'],format="%d/%m/%y")
df['checkin_date']=pd.to_datetime(df['checkin_date'],format="%d/%m/%y")
df['checkout_date']=pd.to_datetime(df['checkout_date'],format="%d/%m/%y")
```

```python
df['season_holidayed_code'].fillna(2,inplace=True)
df['state_code_residence'].fillna(8,inplace=True)
```

```python
df['booking_date_day']=df['booking_date'].dt.day
df['booking_date_month']=df['booking_date'].dt.month
df['booking_date_day_year']=df['booking_date'].dt.year
df['booking_date_dayofweek']=df['booking_date'].dt.dayofweek

df['checkin_date_day']=df['checkin_date'].dt.day
df['checkin_date_month']=df['checkin_date'].dt.month
df['checkin_date_year']=df['checkin_date'].dt.year
df['checkin_date_dayofweek']=df['checkin_date'].dt.dayofweek

df['checkout_date_day']=df['checkout_date'].dt.day
df['checkout_date_month']=df['checkout_date'].dt.month
df['checkout_date_year']=df['checkout_date'].dt.year
df['checkout_date_dayofweek']=df['checkout_date'].dt.dayofweek
```

```python
#to work on
df['booking_date_minus_checkin_date']=(df['booking_date']-df['checkin_date']).dt.days
```

```python
df.drop(['booking_date','checkin_date','checkout_date'],axis=1,inplace=True)
```

```python
df.drop(['memberid'],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le_member_age_buckets=LabelEncoder()
df['member_age_buckets']=le_member_age_buckets.fit_transform(df['member_age_buckets'])

le_cluster_code=LabelEncoder()
df['cluster_code']=le_member_age_buckets.fit_transform(df['cluster_code'])

le_reservationstatusid_code=LabelEncoder()
df['reservationstatusid_code']=le_member_age_buckets.fit_transform(df['reservationstatusid_code'])

le_resort_id=LabelEncoder()
df['resort_id']=le_member_age_buckets.fit_transform(df['resort_id'])

#le_memberid=LabelEncoder()
#df['memberid']=le_member_age_buckets.fit_transform(df['memberid'])
```

```python
from sklearn.preprocessing import OneHotEncoder
one_hot=OneHotEncoder(categorical_features=[0,1,4,5,6,7,9,10,11,13,14,15,16,17,19,23,27,-2,-3],sparse=False)
X_ohe=one_hot.fit_transform(df)
```

```python
df_train=df[df['amount_spent_per_room_night_scaled']!=-999]
df_test=df[df['amount_spent_per_room_night_scaled']==-999]
```

```python
y=df_train['amount_spent_per_room_night_scaled']
X=df_train.drop(['amount_spent_per_room_night_scaled'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=0)
```

```python
ohe_X_train=one_hot.transform(X_train)
```

```python
ohe_X_test=one_hot.transform(X_test)
```

```python
ohe_X_train.shape
```

```python
from catboost import CatBoostRegressor
from sklearn.model_selection import GridSearchCV
cat_boost=CatBoostRegressor(iterations = 10000,
                                task_type = 'GPU',
                               # learning_rate=0.01,
                                eval_metric='RMSE',
                                #depth=8,
                                loss_function= 'RMSE')

cat_boost.fit(ohe_X_train,y_train)
```

```python
y_pred=cat_boost.predict(ohe_X_test)
```

```python
from sklearn.metrics import mean_squared_error
mse_value=mean_squared_error(y_test,y_pred)
rmse_val=np.sqrt(mse_value)
print(rmse_val)
#0.9838598648316379
#0.9813903068065782
#0.9814336183610751
```
